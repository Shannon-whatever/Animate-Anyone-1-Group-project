# DWPose (Test using DWposeDetector)

````python
# Imports
%cd /content/drive/MyDrive/DWPose/ControlNet-v1-1-nightly
import os
import cv2
import matplotlib.pyplot as plt
from annotator.dwpose import DWposeDetector
````

````python
input_dir = '/content/drive/MyDrive/DWPose/ControlNet-v1-1-nightly/test_imgs'
output_dir = '/content/drive/MyDrive/DWPose/ControlNet-v1-1-nightly/test_results'

pose = DWposeDetector()

image_extensions = {'.jpg', '.jpeg', '.png', '.bmp', '.tiff'}
image_files = [f for f in os.listdir(input_dir) if os.path.splitext(f)[1].lower() in image_extensions]

# Process each image
for filename in image_files:
    input_path = os.path.join(input_dir, filename)
    output_path = os.path.join(output_dir, filename)

    oriImg = cv2.imread(input_path)  # B, G, R
    if oriImg is None:
        print(f"Warning: Could not read image {input_path}")
        continue

    out = pose(oriImg)
    plt.imsave(output_path, out)
    print(f"Saved result to {output_path}")

print("Done")
````

# Render DWPose outputs as pose skeleton images similar to OpenPose

We employ DWPose[55] to extract pose sequence of characters in the
video, including **body and hands**, rendering it as pose skeleton images following OpenPose[6].

## OpenPose Keypoints

Body Keypoints (COCO format, 18 points)

````python
0: Nose
1: Left Eye
2: Right Eye
3: Left Ear
4: Right Ear
5: Left Shoulder
6: Right Shoulder
7: Left Elbow
8: Right Elbow
9: Left Wrist
10: Right Wrist
11: Left Hip
12: Right Hip
13: Left Knee
14: Right Knee
15: Left Ankle
16: Right Ankle
17: Neck (sometimes included)
````

Common Skeleton Connections (Pairs of keypoints)

````python
BODY_CONNECTIONS = [
    (0, 1), (0, 2),        # Nose to eyes
    (1, 3), (2, 4),        # Eyes to ears
    (0, 5), (0, 6),        # Nose to shoulders
    (5, 7), (7, 9),        # Left arm
    (6, 8), (8, 10),       # Right arm
    (5, 6),                # Shoulders
    (5, 11), (6, 12),      # Torso
    (11, 13), (13, 15),    # Left leg
    (12, 14), (14, 16),    # Right leg
    (11, 12)               # Hips
]
````

Hand Keypoints (21 points)

````python
HAND_CONNECTIONS = [
    (0, 1), (1, 2), (2, 3), (3, 4),       # Thumb
    (0, 5), (5, 6), (6, 7), (7, 8),       # Index
    (0, 9), (9, 10), (10, 11), (11, 12),  # Middle
    (0, 13), (13, 14), (14, 15), (15, 16),# Ring
    (0, 17), (17, 18), (18, 19), (19, 20) # Pinky
]
````
